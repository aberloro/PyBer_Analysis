```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## Line Chart w Mat Lab Mehtod

```python
# Create the line chart using MatLab
plt.plot(x_axis, y_axis)
```

```python
#formatting w MatLab method

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

## Line Chart w Object Orienterd Interface

```python
# Create the plot with ax.plt() or Objject Oriented Interface

#create a figure
fig, ax = plt.subplots()

#define axes
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() or object oriented interface
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

## Scatter Plots by MatLab Method

```python
import matplotlib.pyplot as plt

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
#adding an 'o'o in the parenthesis 
#makes this a scatter plot vs line plot
#plt.plot(x_axis, y_axis, 'o')

#also possible by plt.scatter() function
plt.scatter(y_axis, x_axis, c="red", marker = "o", label="Chicago")

#format and annottions
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.legend()
plt.gca().invert_yaxis()
plt.xlim(0,50)
```

## Bubble Charts by MatLib Method

```python
#bubble charts add Size 's=' and Color  'c=' factors to scatter charts
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#dots sre too small to read, adjust by x3 on Y values 

#y_axis_larger = []
#for data in y_axis:
    #y_axis_larger.append(data*3)

#and swap in for SIZE parameter    
#plt.scatter(x_axis, y_axis, s=y_axis_larger)


#use list comprehension instead of for and while loops
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

## Scatter Plot Using Object-Oriented Interface

## Bubble Chart in Object Oriented Method

```python
#add the 's' parameter
#make dots bigger with list comprehension feature
fig, ax = plt.subplots()
ax.scatter( y_axis,x_axis, s = [i * 5 for i in y_axis], c="skyblue", alpha=0.80, edgecolor = "k", linewidth = 2, label = "Boston")

ax.legend()
plt.gca().invert_yaxis()
ax.set_xlim(0, 50)
ax.set_ylabel("Month")
ax.set_xlabel("Fare in $")
ax.set_title("PyBer Fare by Month")

plt.show()
#Same in MATLPLOTLIB
#make dots bigger with list comprehension feature
#format color line width, edge, etc in initialization
#plt.scatter( y_axis,x_axis, s = [i * 5 for i in y_axis], c="skyblue", alpha=0.80, edgecolor = "k", linewidth = 2, label = "Boston")

#plt.legend()
#plt.gca().invert_yaxis()
#plt.xlim(0, 50)
#plt.ylabel("Month")
#plt.xlabel("Fare in $")
#plt.title("PyBer Fare by Month")
```

## Pie Charts using MatPlotLib

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#assign a color for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#make it bigger
plt.subplots(figsize=(8,8))

#explode 7th value for visual impact, 0 means no exploding
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#add percentage to wedges
plt.pie(y_axis, 
        explode=explode_values, 
        colors = colors,
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```

## Pie Chart Using Object-Oriented Interface

```python
#set explode list
explode_list = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#colors
color_list = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]


fig, ax = plt.subplots()
ax.pie(y_axis,
       labels=x_axis,
      autopct='%.1f%%',
      explode = explode_list,
      shadow=True,
      startangle=-90,
      colors = color_list)

fig.set_size_inches(8,8)
plt.gca().invert_yaxis()

plt.show()

```
